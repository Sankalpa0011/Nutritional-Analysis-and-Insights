# The data

You source nutrition data from USDA's FoodData Central [website](https://fdc.nal.usda.gov/download-datasets.html). This data contains the calorie content of 7,793 common foods, as well as their nutritional composition. Each row represents one food item, and nutritional values are based on a 100g serving. Here is a description of the columns:

- **FDC_ID**: A unique identifier for each food item in the database.
- **Item**: The name or description of the food product.
- **Category**: The category or classification of the food item, such as "Baked Products" or "Vegetables and Vegetable Products".
- **Calories**: The energy content of the food, presented in kilocalories (kcal).
- **Protein**: The protein content of the food, measured in grams.
- **Carbohydrate**: The carbohydrate content of the food, measured in grams.
- **Total fat**: The total fat content of the food, measured in grams.
- **Cholesterol**: The cholesterol content of the food, measured in milligrams.
- **Fiber**: The dietary fiber content of the food, measured in grams.
- **Water**: The water content of the food, measured in grams.
- **Alcohol**: The alcohol content of the food (if any), measured in grams.
- **Vitamin C**: The Vitamin C content of the food, measured in milligrams.

# Identifying the Fruit with the Highest Vitamin C Content

## Fruit with the Highest Vitamin C Content
To determine the fruit with the highest vitamin C content, we filtered the dataset for items categorized as "Fruit" and sorted the results by the "Vitamin C" column. The analysis revealed that the fruit with the highest vitamin C content is [highest_vit_c_item], containing [highest_vit_c] mg of vitamin C.

```python
import pandas as pd

# Load the dataset
nutrition_df = pd.read_csv('./nutrition.csv')

# Display the first few rows of the dataframe
display(nutrition_df)

# Filter the dataset for fruits and sort by Vitamin C content
df_fruits_vitamin_c = nutrition_df[nutrition_df['Category'].str.contains('Fruit', case=False, na=False)]

# Display filtered dataset
print("\nfruits with Vitamin C content")
display(df_fruits_vitamin_c[['Item', 'Category', 'Vitamin C']])

# Remove 'mg' from 'Vitamin C' column                                           
df_fruits_vitamin_c['Vitamin C'] = df_fruits_vitamin_c['Vitamin C'].str.replace(' mg', '')

# Convert 'Vitamin C' column to float type
df_fruits_vitamin_c['Vitamin C'] = pd.to_numeric(df_fruits_vitamin_c['Vitamin C'].str.replace(' mg', ''), errors='coerce')

# Get the maximum float value from 'Vitamin C' column
highest_vit_c = df_fruits_vitamin_c['Vitamin C'].max()
# Get the highest valued Item
highest_vit_c_index = df_fruits_vitamin_c['Vitamin C'].idxmax()
highest_vit_c_item = df_fruits_vitamin_c.loc[highest_vit_c_index, 'Item']
print(f"* The fruit with the highest Vitamin C content is '{highest_vit_c_item}' and it contains {highest_vit_c} mg of Vitamin C.")
```

# Relationship between Calories and Water Content
## Exploring the Correlation
To understand the relationship between the calories and water content of food items, we created a scatter plot and calculated the correlation coefficient. The analysis showed that [interpretation_of_calories_water_correlation].

```python
# Select relevant columns
calories_water_df = nutrition_df[['Item', 'Category', 'Calories', 'Water']]

# Sort by Calories
calories_water_df = calories_water_df.sort_values(by='Calories', ascending=False)

# Display the DataFrame
display(calories_water_df)



import matplotlib.pyplot as plt

# Scatter plot of calories vs. water content
plt.figure(figsize=(8, 6))
plt.scatter(calories_water_df['Calories'], calories_water_df['Water'], color='blue', alpha=0.5, s=50)
plt.xlabel('Calories (kcal)')
plt.ylabel('Water (g)')
plt.title('Calories vs. Water Content')
plt.show()


# Convert Columns to float values
# print(calories_water_df.dtypes)
calories_water_df['Calories'] = pd.to_numeric(calories_water_df['Calories'].str.replace(' kcal', ''), errors='coerce')
calories_water_df['Water'] = pd.to_numeric(calories_water_df['Water'].str.replace(' g', ''), errors='coerce')
# print(calories_water_df.dtypes)
# display(calories_water_df)

# Correlation analysis
calories_water_corr = calories_water_df['Calories'].corr(calories_water_df['Water'])
print(f"The correlation between calories and water content is: {calories_water_corr:.2f}")
```

### Interpreting the Relationship
The [interpretation_of_calories_water_correlation] suggests that [further_interpretation_of_relationship].

```python
# Interpret the relashionship
if calories_water_corr < 0:
    print("There is negative correlation between calories and water content. Foods with higher water content tend to have fewer calories")
elif calories_water_corr > 0:
    print("There is positive correlation between calories and water content. Foods with higher water content tend to have more calories")
else: 
    print("There is no linear correlation between calories and water content")
```

# Drawbacks of Extreme Dietary Approaches
## Drawbacks of a Zero-Carb Diet
- Nutrient Deficiencies
- Gut Health Issues
- Energy and Fatigue
- Long-term Sustainability
## Drawbacks of a Very High-Protein Diet
- Kidney Strain
- Calcium Loss
- Dehydration
- Digestive Issues
- Liver Stress

```python
# Drawbacks of a Zero-Carb Diet
def zero_carb_diet_drawbacks():
    """
    Discusses the potential drawbacks of a zero-carb diet.
    """
    print("Potential Drawbacks of a Zero-Carb Diet:")
    print("1. Nutrient Deficiencies:")
    print("   - Carbohydrates are the primary source of many essential vitamins, minerals, and fiber.")
    print("   - Eliminating carbohydrates entirely can lead to deficiencies in these vital nutrients.")
    print("2. Gut Health Issues:")
    print("   - Carbohydrates, especially fiber-rich sources, are crucial for maintaining a healthy gut microbiome.")
    print("   - A zero-carb diet may disrupt the balance of gut bacteria, leading to digestive problems.")
    print("3. Energy and Fatigue:")
    print("   - Carbohydrates are the body's primary source of energy. Eliminating them can result in decreased energy levels and fatigue.")
    print("4. Long-term Sustainability:")
    print("   - Maintaining a zero-carb diet in the long term can be challenging and may lead to poor adherence.")
    print("   - It may be difficult to meet all nutritional needs without any carbohydrate intake.")

# Drawbacks of a Very High-Protein Diet
def high_protein_diet_drawbacks():
    """
    Discusses the potential drawbacks of a very high-protein diet.
    """
    print("Potential Drawbacks of a Very High-Protein Diet:")
    print("1. Kidney Strain:")
    print("   - The metabolism of excess protein can place a significant burden on the kidneys.")
    print("   - This may lead to an increased risk of kidney stones and other kidney-related issues.")
    print("2. Calcium Loss:")
    print("   - A high-protein diet can increase the excretion of calcium, potentially leading to weaker bones and an increased risk of osteoporosis.")
    print("3. Dehydration:")
    print("   - The metabolism of protein can lead to increased water loss, which can result in dehydration if not properly managed.")
    print("4. Digestive Issues:")
    print("   - Consuming very high amounts of protein may cause digestive problems, such as bloating, constipation, or diarrhea.")
    print("5. Liver Stress:")
    print("   - The liver is responsible for processing and breaking down excess protein, and a very high-protein diet can potentially overload the liver's capacity.")

# Example usage
zero_carb_diet_drawbacks()
print("\n")
high_protein_diet_drawbacks()
```

# Evaluating the Linear Model and Analyzing Errors
## Fitting the Linear Model
We fitted a linear model to the data to test whether the estimates from the Cleveland Clinic website align with the observed relationships in the dataset.

```python
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

# Create dataframe
model_df = nutrition_df[['Item', 'Protein', 'Carbohydrate', 'Total fat', 'Calories']]
display(model_df)

# Convert column values to numeric
#model_df['Protein'] = pd.to_numeric(model_df['Protein'].str.replace(' g', ''), errors='coerce')
#model_df['Carbohydrate'] = pd.to_numeric(model_df['Carbohydrate'].str.replace(' g', ''), errors='coerce')
#model_df['Total fat'] = pd.to_numeric(model_df['Total fat'].str.replace(' g', ''), errors='coerce')
#model_df['Calories'] = pd.to_numeric(model_df['Calories'].str.replace(' kcal', ''), errors='coerce')

model_df.loc[:, 'Protein'] = pd.to_numeric(model_df['Protein'].str.replace(' g', ''), errors='coerce')
model_df.loc[:, 'Carbohydrate'] = pd.to_numeric(model_df['Carbohydrate'].str.replace(' g', ''), errors='coerce')
model_df.loc[:, 'Total fat'] = pd.to_numeric(model_df['Total fat'].str.replace(' g', ''), errors='coerce')
model_df.loc[:, 'Calories'] = pd.to_numeric(model_df['Calories'].str.replace(' kcal', ''), errors='coerce')


# Prepare the data
X = model_df[['Protein', 'Carbohydrate', 'Total fat']]
y = model_df['Calories']

# Fit the linear model
model = LinearRegression()
model.fit(X, y)

# Print the model coefficient
print("Model coefficients")
print(f"-Intercept: {model.intercept_:.2f}")
print(f"-Protein coeffficient: {model.coef_[0]:.2f}")
print(f"-Carbohydrate coefficient: {model.coef_[1]:.2f}")
print(f"-Total fat coefficient: {model.coef_[2]:.2f}")
```

## Comparison to Cleveland Clinic Estimates
The model coefficients for [protein/carbohydrate/fat] were [protein_coefficient:.2f]/[carbohydrate_coefficient:.2f]/[total_fat_coefficient:.2f], which [comparison_to_cleveland_clinic_estimates] the Cleveland Clinic estimates of 4 kcal/g for protein and carbohydrate, and 9 kcal/g for fat.

```python
# Compare the coefficients with estimates from the cleveland Clinic website
print("\nCleveland clinic estimates")
print("-Protein: 4 kcal/g")
print("-Carbohydrate: 4 kcal/g")
print("-Total fat: 9 kcal/g")

# Analyze the model errors
residuals = y - model.predict(X)
print(f"\nRoot Mean Squared Error: {np.sqrt(np.mean(residuals**2)):.2f}")
```

# Analyzing the Residuals

```python
# Analyze the model errors
residuals = y - model.predict(X)

# Calculate the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(residuals**2))
print(f"Root Mean Squared Error: {rmse:.2f}")

# Inspect the distribution of residuals
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=20)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Model Residuals')
plt.show()

# Look for patterns in the residuals
plt.figure(figsize=(8, 6))
plt.scatter(model_df['Calories'], residuals)
plt.xlabel('Actual Calories')
plt.ylabel('Residuals')
plt.title('Residuals vs Actual Calories')
plt.show()
```
